```python
import json
from copy import deepcopy
```

```python
public_test_2_path = "../../data/VLSP 2025 - MLQA-TSR Data Release/public_test/vlsp_2025_public_test_task2.json"
with open(public_test_2_path, "r") as f:
    public_test = json.load(f)
```

```python
submission_results = []
for record in public_test:
    submission_result = deepcopy(record)
    if submission_result["question_type"] == "Multiple choice":
        submission_result["answer"] = "A"
    else:
        submission_result["answer"] = "Yes"
    submission_results.append(submission_result)
```

```python
submission_results
```

```python
result_file = "../../submission_result/submission_task2.json"
with open(result_file, "w") as f:
    json.dump(submission_results, f, indent=2)
```
